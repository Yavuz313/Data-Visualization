## Kütüphaneleri import edin

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings;
warnings.filterwarnings("ignore")
```

## Değişkenlerimizi tanımlayalım

m değerini output a göre numpy kütüphanesi yardımı ile oluşturalım

```python
m = np.arange(0,11, dtype=float)
```

```python
m
```

```python
c = 3* 10**8
```

```python
E = m* c**2
```

```python
E
```

## Oluşturduğumuz dokümanları plotlayalım

#### PLOTTING E=mc^2

```python
plt.plot(m, E, "orange")

plt.xlabel("M in g")
plt.ylabel("E in J")

plt.title("E=mc^2")

plt.show()
```

## Alttaki görseli oluştururken farklı bir değişken tanımlamayın. Bence eksenlerden birinin logaritmik hesaplanması gerekebilir:)

## Aşağıdaki değişkenlere göre ilgili platlamaları yapın

```python
labels = ["1 Mo", "3 Mo", "6 Mo", "1 Yr", "2 Yr", "3 Yr", "5 Yr", "7 Yr", "10 Yr", "20 Yr", "30 Yr"]
july16_2007 = [4.75, 4.98, 5.08, 5.01, 4.89, 4.89, 4.95, 4.99, 5.05, 5.21, 5.14]
july16_2020 = [0.12, 0.11, 0.13, 0.14, 0.16, 0.17, 0.28, 0.46, 0.62, 1.09, 1.31]
```

```python
plt.figure(figsize = (7, 5))

plt.plot(labels, july16_2007, "b", label="july16_2007")
plt.plot(labels, july16_2020, "orange", label="july16_2020")

plt.legend()

plt.show()
```

```python
plt.figure(figsize = (7, 5))

plt.plot(labels, july16_2007, "b", label="july16_2007")
plt.plot(labels, july16_2020, "orange", label="july16_2020")

plt.legend(bbox_to_anchor=(1, 0.6))

plt.show()
```

```python
plt.figure(figsize = (7, 5))

plt.plot(labels, july16_2007, "b", label="july16_2007")
plt.plot(labels, july16_2020, "orange", label="july16_2020")

plt.legend(bbox_to_anchor=(0.6, 0.1, 0.7, 0.1))

plt.show()
```

## Alttaki görselleri araştırarak yapmaya çalışalım. Zihnimizi ve parmaklarımızı zorlayalım:)

#### Bir hatırlatma "fig, ax = plt.subplots" syntax i size cevabı bulmanızda yardımcı olur. 

#### ayrıca shift + tab+ tab yapmayı unutmayın ve içindeki parametreleri denemeyi unutmayın.

```python
plt.figure(figsize = (7, 5))

plt.subplot(2, 1, 1) 
plt.plot(labels, july16_2007, "b")

plt.subplot(2, 1, 2)
plt.plot(labels, july16_2020, "b")


plt.show()
```

## Bu görselleri alabilenler için alkışı borç bilirim

```python

fig, ax1 = plt.subplots()

ax1.plot(labels, july16_2007, 'b-', label='July 16 2007')
ax1.set_xlabel('Maturity')
ax1.set_ylabel('July 16 2007', color='b')
ax1.tick_params(axis='y', labelcolor='b')

ax2 = ax1.twinx()
ax2.plot(labels, july16_2020, 'r-', label='July 16 2020')
ax2.set_ylabel('July 16 2020', color='r')
ax2.tick_params(axis='y', labelcolor='r')

plt.show()
```

```python

fig, ax1 = plt.subplots()

ax1.plot(labels, july16_2007, 'b-', label='July 16 2007')
ax1.set_xlabel('Maturity')
ax1.set_ylabel('July 16 2007', color='b')
ax1.tick_params(axis='y', labelcolor='b')
ax1.spines['left'].set_color('blue')  
ax1.spines['left'].set_linewidth(2)   

ax2 = ax1.twinx()
ax2.plot(labels, july16_2020, 'r-', label='July 16 2020')
ax2.set_ylabel('July 16 2020', color='r')
ax2.tick_params(axis='y', labelcolor='r')
ax2.spines['right'].set_color('red') 
ax2.spines['right'].set_linewidth(2) 

plt.show()
```
